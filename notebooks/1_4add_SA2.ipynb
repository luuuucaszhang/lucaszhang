```python
from shapely.geometry import Point, Polygon
import pandas as pd
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
geopackage_path = "../data/landing/Geopackage/LGA ERP GeoPackage 2022.gpkg"
gdf = gpd.read_file(geopackage_path)
```

```python
def add_SA2(input_file, out_file):
# adding the property's LGA code name and area in to the csv
# input_file: the csv file needed to add in more features 
# output_file: is the csv name and the output path of the merged/adjusted csv file
    geopackage_path = "../data/landing/Geopackage/SA2 ERP GeoPackage 2022.gpkg"
    gdf = gpd.read_file(geopackage_path)
    selected_row = gdf[gdf["State_name_2021"]=="Victoria"]
    data = pd.read_csv(input_file)
    
    sa2_code_list=[]
    sa2_name_list = []
    
    for count in range(len(data)):
        point = Point(data["longitude"][count],data["Latitude"][count])
        matched=False 
        for count in range(len(selected_row)):
            polygon=selected_row["geometry"].iloc[count]
            is_inside = polygon.contains(point)
            if is_inside:
                matched=True
                sa2_name = selected_row["SA2_name_2021"].iloc[count]
                sa2_code = selected_row["SA2_code_2021"].iloc[count]

                break
            
        if matched:
            sa2_code_list.append(sa2_code)
            sa2_name_list.append(sa2_name)
                
        else:
            sa2_name_list.append(None)
            sa2_code_list.append(None)
       

    data["SA2_code"]=sa2_code_list
    data["SA2_name"]=sa2_name_list
    if not os.path.exists("../data/curated/second_clean"):
        os.makedirs("../data/curated/second_clean")
        
    data.to_csv(out_file,index=False)
```

```python
add_SA2("../data/curated/first_clean/merge_shopping_center_apartment1.csv","../data/curated/second_clean/apartment_1.csv")
add_SA2("../data/curated/first_clean/merge_shopping_center_apartment2.csv","../data/curated/second_clean/apartment_2.csv")
add_SA2("../data/curated/first_clean/merge_shopping_center_apartment3.csv","../data/curated/second_clean/apartment_3.csv")
add_SA2("../data/curated/first_clean/merge_shopping_center_apartment4.csv","../data/curated/second_clean/apartment_4.csv")
add_SA2("../data/curated/first_clean/merge_shopping_center_apartment5.csv","../data/curated/second_clean/apartment_5.csv")
add_SA2("../data/curated/first_clean/merge_shopping_center_house_1.csv","../data/curated/second_clean/house_1.csv")  
add_SA2("../data/curated/first_clean/merge_shopping_center_house_2.csv","../data/curated/second_clean/house_2.csv")
add_SA2("../data/curated/first_clean/merge_shopping_center_house_3.csv","../data/curated/second_clean/house_3.csv")  
add_SA2("../data/curated/first_clean/merge_shopping_center_house_4.csv","../data/curated/second_clean/house_4.csv")  
add_SA2("../data/curated/first_clean/merge_shopping_center_town_house.csv","../data/curated/second_clean/town_house.csv")
```
