```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'
}


url = 'https://police-station.com/list-of-victoria-police-stations-in-australia/'
response = requests.get(url, headers=header)

soup = BeautifulSoup(response.text, 'html.parser')


table_element = soup.find('table', {'class': 'tablepress tablepress-id-23'})


addresses = []

if table_element:
    column_2_elements = table_element.find_all(class_='column-2')
    
    for element in column_2_elements:
        address = element.get_text(strip=True)
        addresses.append(address)


    for address in addresses:
        print(address)
else:
    print("Unsuccessful")

df = pd.DataFrame(addresses[1:], columns=[addresses[0]])
df.to_csv("../data/rawpolicastation_addresses.csv", index=False)


```
