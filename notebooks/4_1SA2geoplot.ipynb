```python
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
import folium
geopackage_path = "../data/landing/Geopackage/SA2 ERP GeoPackage 2022.gpkg"
gdf = gpd.read_file(geopackage_path)
```

```python
selected_row = gdf[gdf["State_name_2021"]=="Victoria"]
selected_rowq2= selected_row [selected_row["SA2_code_2021"].isin ([205041096,205051101,204021063,201011006,204031492,202011021,216031416,205041095,214021378,214021379,202011022])]
selected_rowq2=selected_rowq2.reset_index(drop=True)
value_list = [22.261132,19.613166,18.20991,23.923410, 20.371659,19.249479,30.844104,26.835844,18.561291,18.561291,19.272195]
selected_rowq2["value"]= value_list
```

```python
geoJSONq2 = selected_rowq2[['SA2_name_2021', 'geometry']].drop_duplicates('SA2_name_2021').to_json()
selected_rowq2['centroid'] = selected_rowq2['geometry'].apply(lambda x: (x.centroid.y, x.centroid.x))
```

```python
m = folium.Map(location=[-37.8136, 144.9631], tiles="Stamen Terrain", zoom_start=8)

# refer to the folium documentations on more information on how to plot aggregated data.
c = folium.Choropleth(
    geo_data=geoJSONq2, # geoJSON 
    name='choropleth', # name of plot
    data=selected_rowq2.reset_index(), # data source
    columns=['SA2_name_2021','value'], # the columns required
    key_on='properties.SA2_name_2021', # this is from the geoJSON's properties
    fill_color='RdYlBu', # color scheme
    nan_fill_color='black',
    legend_name='Rental Price growth rate'
)

c.add_to(m)

m
```

```python
import folium
import geopy.distance

# 初始坐标
coord = [-37.8136, 144.9631]
map_center = coord

# 创建地图
m = folium.Map(location=map_center, tiles="Stamen Terrain", zoom_start=14)

# 计算2公里的经纬度范围
radius_km = 2
radius = geopy.distance.distance(kilometers=radius_km)
sw_coord = radius.destination(point=coord, bearing=225)  # 计算西南角坐标
ne_coord = radius.destination(point=coord, bearing=45)   # 计算东北角坐标

# 创建阴影区域
folium.Circle(
    location=coord,
    radius=2000,  # 半径为2公里，单位为米
    color='gray',
    fill=True,
    fill_opacity=0.2,
).add_to(m)

# 在坐标点上添加标记
folium.Marker(
    location=coord,
    icon=folium.Icon(color='blue'),
).add_to(m)

# 显示地图
m
```
