```python
import requests
import openpyxl
from openpyxl.utils.dataframe import dataframe_to_rows
import pandas as pd
```

```python
url = "https://www.dffh.vic.gov.au/moving-annual-rents-suburb-september-quarter-2022-excel"

response = requests.get(url)

if response.status_code == 200:
    with open("../data/landing/moving_annual_rents_september_2022.xlsx", "wb") as file:
        file.write(response.content)

    print("XLSX file downloaded successfully.")

else:
    print(f"Failed to download the XLSX file. HTTP status code: {response.status_code}")
```

```python
annual_rent_df = pd.read_excel('../data/landing/moving_annual_rents_september_2022.xlsx', sheet_name='All properties')
annual_rent_df
```

```python
# Use iloc to slice the DataFrame and exclude the first two rows and the first column
filtered_df = annual_rent_df.iloc[2:, 1:]

filtered_df
```

```python
start_column_index = 4

# Create a list of column indices to extract
column_indices = [start_column_index + 8 * i for i in range(2022 - 2000 + 1)]
column_indices = [0] + column_indices 

# Extract the columns of the median of annual rent for each year of every suburb
filtered_df = filtered_df.iloc[:, column_indices]
filtered_df
```

```python
# Create a list of new column names
new_column_names = ['suburb'] + [str(year) for year in range(2000, 2023)]

# Use the .rename method to set the new column names
filtered_df.rename(columns=dict(zip(filtered_df.columns, new_column_names)), inplace=True)

filtered_df = filtered_df[filtered_df['suburb'] != 'Group Total']

filtered_df= filtered_df.drop(158) # drop the suburb that does not exist in other datasets

filtered_df.reset_index(drop=True, inplace=True)

filtered_df.to_csv('../data/raw/time_series_annual_rent.csv', index= False)


```
