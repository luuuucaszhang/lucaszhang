```python
import os
import openpyxl
import re
import csv
import os
import pandas as pd
def add_wealth(input_file, output_file):
# add to each property the income median of this property's sa2 area 
# input_file: the csv file needed to add in more features 
# output_file: is the csv name and the output path of the merged/adjusted csv file
# however this is not used after all
    workbook = openpyxl.load_workbook('../data/landing/wealth_data/14100DO0004_2011-22.xlsx')
    input_data = pd.read_csv(input_file)
    sheet=workbook["Table 1"]
    pattern = re.compile(r'2\d{8}$')
    matching_rows = []
    row_iterator= sheet.iter_rows(values_only=True)
    for i in range(6):
        next (row_iterator)
    for i in row_iterator:
        next(row_iterator)
        break
    for row in row_iterator:
        code_value = str(row[0])
        if pattern.match(code_value):
            if str(row[2])=="2019":
                matching_rows.append(row)
                
    value_list=[]
    for i in matching_rows:
        value_list.append([i[0],i[6],i[7],i[30],i[31]]) 
    columns=["SA2_code","Total_employee_income_median","Total_employee_income_mean","Total_income_median","Total_income_mean"]
    df = pd.DataFrame(value_list, columns=columns)
    
    value_1_list=[]
    value_2_list=[]
    value_3_list=[]
    value_4_list=[]
    value_5_list=[]
    # each property 
    for count in range(len(input_data)):
        
        sa2_code = input_data["SA2_code"][count]
        
        matched=False 
        
        for value_count in range(len(df)):
            value_sa2_code = df["SA2_code"][value_count]
            if sa2_code==value_sa2_code:
                matched= True
                value_1= df["Total_employee_income_median"][value_count]
                value_2=df["Total_employee_income_mean"][value_count]
                value_3= df["Total_income_median"][value_count]
                value_4= df["Total_income_mean"][value_count]
            if matched:
                if value_1=="-":
                    value_1_list.append(None)
                else:
                    value_1_list.append(value_1)
                if value_2=="-":
                    value_2_list.append(None)
                else:
                    value_2_list.append(value_2)
                if value_3=="-":
                    value_3_list.append(None)
                else:
                    value_3_list.append(value_3)
                if value_4=="-":
                    value_4_list.append(None)
                else:
                    value_4_list.append(value_4)
                break

        if not matched:
            value_1_list.append(None)
            value_2_list.append(None)
            value_3_list.append(None)
            value_4_list.append(None)
                
    
    input_data["Total_employee_income_median"]=value_1_list
    input_data["Total_employee_income_mean"]=value_2_list
    input_data["Total_income_median"]=value_3_list
    input_data["Total_income_mean"]=value_4_list
    if not os.path.exists("../data/curated/sixth_clean/"):
        os.makedirs("../data/curated/sixth_clean/")
    input_data.to_csv(output_file,index=False)
```

```python
add_wealth("../data/curated/fifth_clean/apartment_1.csv","../data/curated/sixth_clean/apartment_1.csv")
add_wealth("../data/curated/fifth_clean/apartment_2.csv","../data/curated/sixth_clean/apartment_2.csv")
add_wealth("../data/curated/fifth_clean/apartment_3.csv","../data/curated/sixth_clean/apartment_3.csv")
add_wealth("../data/curated/fifth_clean/apartment_4.csv","../data/curated/sixth_clean/apartment_4.csv")
add_wealth("../data/curated/fifth_clean/apartment_5.csv","../data/curated/sixth_clean/apartment_5.csv")
add_wealth("../data/curated/fifth_clean/house_1.csv","../data/curated/sixth_clean/house_1.csv")  
add_wealth("../data/curated/fifth_clean/house_2.csv","../data/curated/sixth_clean/house_2.csv")  
add_wealth("../data/curated/fifth_clean/house_3.csv","../data/curated/sixth_clean/house_3.csv")  
add_wealth("../data/curated/fifth_clean/house_4.csv","../data/curated/sixth_clean/house_4.csv")  
add_wealth("../data/curated/fifth_clean/town_house.csv","../data/curated/sixth_clean/town_house.csv")
```
